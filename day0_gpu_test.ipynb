```python
import torch, time

x = torch.randn(4096, 4096)
y = torch.randn(4096, 4096)

t0 = time.time()
for _ in range(10):
    _ = x @ y
cpu_t = time.time() - t0

xg, yg = x.cuda(), y.cuda()
torch.cuda.synchronize()
t0 = time.time()
for _ in range(10):
    _ = xg @ yg
torch.cuda.synchronize()
gpu_t = time.time() - t0

print(f"CPU (Ultra 7 155H): {cpu_t:.3f} s")
print(f"GPU (RTX 4060):     {gpu_t:.3f} s")
print(f"GPU is {cpu_t/gpu_t:.0f}x faster")
```
