```python
import numpy as np
import matplotlib.pyplot as plt

x_data = [1.0, 2.0, 3.0]
y_data = [3.0, 5.0, 7.0]


w_list = []
mse_list = []
b_list = []

w_range = np.arange(0.0, 4.1, 0.1)
b_range = np.arange(-2.0, 2.1, 0.1)

# build the 2d grids
W, B = np.meshgrid(w_range, b_range)

# accumulate cost over the 3 data points
cost = np.zeros_like(W)

for x_val, y_val in zip(x_data, y_data):
    y_pred_val = x_val * W + B
    loss_val = (y_pred_val - y_val) * (y_pred_val - y_val)
    cost += loss_val

cost = cost / len(x_data)

# plot
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.plot_surface(W, B, cost, cmap = 'viridis')
ax.set_xlabel('w')
ax.set_ylabel('b')
ax.set_zlabel('cost')
plt.show()
```
